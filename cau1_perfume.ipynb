```python
from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt
```

```python
stopwords=set(STOPWORDS)
```

```python
f=open('Du_lieu_cung_cap/text2.txt')
text=f.read()
text=text.replace('\n',' ')

wc=WordCloud(background_color='black',max_words=100,stopwords=stopwords)
wc.generate(text)

plt.figure(figsize=(10,12))
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import numpy as np
from PIL import Image
wc_mask=np.array(Image.open('Du_lieu_cung_cap/perfume.png'))
wc=WordCloud(background_color='white',mask=wc_mask,max_words=100,stopwords=stopwords)
wc.generate(text)

plt.figure(figsize=(10,12))
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```
